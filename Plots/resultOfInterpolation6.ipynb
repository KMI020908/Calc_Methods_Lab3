Экстрополирование экспоненты на равномерной сетке [0 ; 2]

```julia
grid = readdlm("D:\\Calc_Methods\\Lab3\\exponentExtropolation.txt")
intervalX = [grid[i] for i in 1 : size(grid)[1]]
intervalY = [grid[i] for i in size(grid)[1] + 1 : length(grid)]
layout = Layout(
        font_family = "Times New Roman",
        font_color = "black",
        title = attr(
            paper_bgcolor = "red",
            text = "Экстрополирование eˣ с равномерной сеткой на [0, 2]. n = 10",
            font_family = "Times New Roman",
            font_color = "black",
            font_size = 24,
            autosize = false
        ),
        height = 600,
        width = 800,
        paper_bgcolor = "white",
        plot_bgcolor = "white",
        coloraxis = attr(
          outlinecolor = "black"  
        ),
        xaxis = attr(
            title = attr(
                text = "x",
                font_family = "Times New Roman",
                font_color = "black",
                font_size = 20,
            ),
            color = "black",
            gridcolor = "black",
            gridwidth = 1,
            zerolinecolor = "black",
            zerolinewidth = 3,
            tickfont = attr(
                size = 15
            )
        ),
        yaxis = attr(
            title = attr(
                text = "y",
                font_family = "Times New Roman",
                font_color = "black",
                font_size = 20,
            ),
            gridcolor = "black",
            exponentformat = "power",
            gridwidth = 1 ,
            zerolinecolor = "black",
            zerolinewidth = 3,
            tickfont = attr(
                size = 15
            )   
        )
    )

    lineAttr = attr(
        color = funcColor,
        width = 3
    )

    pl = plot(
        [scatter(x = intervalX, y =intervalY, mode = "line", line = lineAttr, name = "Lₙ(x)"),
        scatter(x = intervalX, y =exp.(intervalX), mode = "line", line = attr(color = "black",
        width = 3), name = "eˣ")],
        layout
    )
```
